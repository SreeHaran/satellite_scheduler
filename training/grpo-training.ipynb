```python
import textwrap

SYSTEM_PROMPT = textwrap.dedent(
    """
    You are an AI agent controlling a satellite for mission-critical imaging and data downlink. You will Capture priority targets, compress data, downlink to ground station, and manage battery.

    Mission objective:
    - maximize completed mission value
    - prefer high-priority requests
    - capture useful images before deadlines
    - compress raw data before downlink
    - downlink compressed data during ground-station visibility
    - preserve battery and avoid storage overflow

    Allowed actions:
    1. wait
    2. abort_task
    3. sun_point_for_charging
    4. compress_data
    5. downlink_to_station
    6. capture_image

    Important environment facts:
    - capture_image requires a target_id
    - capture creates RAW data
    - compress_data converts RAW data into COMPRESSED data and frees storage
    - downlink_to_station only helps when ground station is visible and compressed data exists
    - sun_point_for_charging only helps when sunlit
    - if the satellite is already busy, any non-abort action continues the current task

    Response rules:
    - output exactly one JSON object
    - no markdown
    - no prose
    - schema:
        - For actions without parameters: "wait", "abort_task", "sun_point_for_charging", "compress_data", "downlink_to_station"
        - For capture_image action: "capture_image <target-id>". Example: "capture_image 42"
    """
).strip()
```

```python
from typing import Any, Dict, List, Optional, Tuple

def build_user_prompt(
    step: int,
    obs: SatelliteSchedulerObservation,
    last_reward: float,
    history: List[str],
) -> str:
    # Extract all state attributes from observation
    current_time = obs.current_time
    attitude = obs.attitude
    busy_status = obs.busy_status
    remaining_action_steps = obs.remaining_action_steps
    battery_level = obs.battery_level
    storage_used = obs.storage_used
    raw_data_amount = obs.raw_data_amount
    compressed_data_amount = obs.compressed_data_amount
    sunlit_status = obs.sunlit_status
    ground_station_visible = obs.ground_station_visible
    pending_request_queue = obs.pending_request_queue
    current_selected_request_id = obs.current_selected_request_id

    sunlit_str = "SUNLIT" if sunlit_status else "DARK"
    gs_str = "VISIBLE" if ground_station_visible else "NOT VISIBLE"

    try:
        if pending_request_queue:
            queue_lines = "\n".join(
                f"  - id={r.request_id} priority={r.priority.value} mode={r.imaging_mode.value} deadline={r.deadline}s"
                for r in pending_request_queue
            )
            queue_str = f"{len(pending_request_queue)} pending:\n{queue_lines}"
        else:
            queue_str = "none"
    except Exception as e:
        queue_str = "(error reading queue)"

    return textwrap.dedent(
        f"""
        SATELLITE STATE
        ===============
        Step: {step}/{MAX_STEPS}
        
        ORBITAL STATE
        Time: {current_time}s | Attitude: {attitude} | Busy Status: {busy_status}
        Remaining Action Steps: {remaining_action_steps}
        
        RESOURCE LEVELS
        Battery: {battery_level:.1f}% | Storage Used: {storage_used:.1f} GB
        Raw Data: {raw_data_amount:.1f} GB | Compressed Data: {compressed_data_amount:.1f} GB
        
        VISIBILITY
        Sunlit: {sunlit_str} | Ground Station: {gs_str}
        
        MISSION REQUESTS
        {queue_str}
        Currently Selected Request: {current_selected_request_id if current_selected_request_id else "None"}
        
        PERFORMANCE
        Last Reward: {last_reward:.3f}
        
        RECENT ACTIONS
        {chr(10).join(history[-3:]) if history else "(none)"}
        
        What action should the satellite take next?
    """
    ).strip()

def validate_action(action: Dict[str, Any]) -> Dict[str, Any]:
    allowed = {
        "wait",
        "abort_task",
        "sun_point_for_charging",
        "compress_data",
        "downlink_to_station",
        "capture_image",
    }

    action_type = action.get("action_type")
    if action_type not in allowed:
        raise ValueError(f"Invalid action_type: {action_type}")

    if action_type == "capture_image":
        target_id = action.get("target_id")
        if target_id is None or str(target_id).strip() == "":
            raise ValueError("capture_image requires target_id")
        return {"action_type": "capture_image", "target_id": str(target_id)}

    return {"action_type": action_type}


def parse_action(action: Dict[str, Any]) -> Optional[SatelliteSchedulerAction]:
    """Parse LLM response into a SatelliteSchedulerAction."""
    try:
        text_lower = action.action_type

        # Handle capture_image first: extract target_id from response (e.g. "capture_image 42")
        if action_type == "capture_image":
            match = re.search(r"\d+", text)
            target_id = int(match.group()) if match else None
            return SatelliteSchedulerAction(
                action_type=ActionType.CAPTURE_IMAGE, target_id=target_id
            )

        action_mappings = {
            "wait": ActionType.WAIT,
            "abort": ActionType.ABORT_TASK,
            "sun_point": ActionType.SUN_POINT_FOR_CHARGING,
            "charge": ActionType.SUN_POINT_FOR_CHARGING,
            "compress": ActionType.COMPRESS_DATA,
            "downlink": ActionType.DOWNLINK_TO_STATION,
        }

        for keyword, action_type in action_mappings.items():
            if keyword in text_lower:
                return SatelliteSchedulerAction(action_type=action_type)

        # Default to wait if unclear
        return SatelliteSchedulerAction(action_type=ActionType.WAIT)
    except Exception as e:
        return SatelliteSchedulerAction(action_type=ActionType.WAIT)
```

```python
import json
from datasets import Dataset

def build_grpo_dataset(data_path="/kaggle/working/satellite_scheduler/dataset/grpo_training.jsonl", sample_size=500):
    """
    Load real satellite state observations from SFT data as GRPO prompts.
    Each prompt is a chat-formatted message with the system prompt + a real
    satellite state, so the model must choose the best action for that state.
    """
    rows = []
    with open(data_path) as f:
        for line in f:
            d = json.loads(line)
            rows.append({
                "prompt": [
                    {"role": "system", "content": d["system"]},
                    {"role": "user",   "content": d["prompt"]},
                ],
            })
    ds = Dataset.from_list(rows)
    if sample_size and sample_size < len(ds):
        ds = ds.shuffle(seed=42).select(range(sample_size))
    print(f"GRPO dataset: {len(ds)} diverse satellite states")
    return ds
```

```python
import re


def parse_action_from_text(text: str) -> SatelliteSchedulerAction:
    """Parse raw LLM text into a SatelliteSchedulerAction."""
    try:
        text_lower = text.lower().strip()
        if "capture" in text_lower or "image" in text_lower:
            match = re.search(r"\d+", text)
            target_id = int(match.group()) if match else None
            return SatelliteSchedulerAction(
                action_type=ActionType.CAPTURE_IMAGE, target_id=target_id
            )
        for kw, at in {
            "wait": ActionType.WAIT, "abort": ActionType.ABORT_TASK,
            "sun_point": ActionType.SUN_POINT_FOR_CHARGING, "charge": ActionType.SUN_POINT_FOR_CHARGING,
            "compress": ActionType.COMPRESS_DATA, "downlink": ActionType.DOWNLINK_TO_STATION,
        }.items():
            if kw in text_lower:
                return SatelliteSchedulerAction(action_type=at)
        return SatelliteSchedulerAction(action_type=ActionType.WAIT)
    except Exception:
        return SatelliteSchedulerAction(action_type=ActionType.WAIT)


def _parse_state(prompt):
    """Extract satellite state values from the prompt text."""
    if isinstance(prompt, list):
        text = prompt[-1]["content"] if prompt else ""
    else:
        text = str(prompt)

    def _ext(pattern, default=None):
        m = re.search(pattern, text, re.IGNORECASE)
        return m.group(1) if m else default

    state = {}
    state["battery"] = float(_ext(r"Battery Level:\s*([\d.]+)", "50"))
    state["storage"] = float(_ext(r"Storage Used:\s*([\d.]+)", "0"))
    state["raw_data"] = float(_ext(r"Raw Data Amount:\s*([\d.]+)", "0"))
    state["compressed_data"] = float(_ext(r"Compressed Data Amount:\s*([\d.]+)", "0"))

    sunlit_raw = _ext(r"Sunlit:\s*(True|False|SUNLIT|DARK)", "False")
    state["sunlit"] = sunlit_raw.lower() in ("true", "sunlit")

    gs_raw = _ext(r"Ground Station Visible:\s*(True|False|VISIBLE|NOT VISIBLE)", "False")
    state["gs_visible"] = gs_raw.lower() in ("true", "visible")

    state["busy"] = _ext(r"Busy Status:\s*(\w+)", "idle")
    state["remaining_steps"] = int(_ext(r"Remaining Action Steps:\s*(\d+)", "0"))

    requests = []
    for m in re.finditer(r"id=(\d+)\s+priority=(\w+)", text):
        requests.append({"id": int(m.group(1)), "priority": m.group(2).lower()})
    state["requests"] = requests

    return state


def _get_completion_text(c):
    if isinstance(c, list):
        return c[-1]["content"] if c else ""
    if isinstance(c, dict):
        return c.get("content", str(c))
    return str(c)


# ── Reward 1: Format correctness ─────────────────────────────────
def reward_format(completions, **kwargs):
    """Reward clean, concise action output."""
    rewards = []
    for c in completions:
        text = _get_completion_text(c).strip()
        action = parse_action_from_text(text)
        words = text.split()

        if action.action_type == ActionType.CAPTURE_IMAGE:
            # "capture_image 42" → 2 tokens is perfect
            if len(words) <= 3 and action.target_id is not None:
                rewards.append(1.0)
            elif action.target_id is not None:
                rewards.append(0.6)
            else:
                rewards.append(0.2)
        elif len(words) <= 2:
            rewards.append(0.9)   # Clean single-word action
        elif len(words) <= 5:
            rewards.append(0.1)   # Parsable but verbose
        else:
            rewards.append(-0.5)   # Too much extra text
    return rewards


# ── Reward 2: Contextual appropriateness ──────────────────────────
def reward_context(completions, prompts, **kwargs):
    """Score how well the action fits the current satellite state."""
    rewards = []
    for c, p in zip(completions, prompts):
        text = _get_completion_text(c).strip()
        action = parse_action_from_text(text)
        state = _parse_state(p)
        at = action.action_type

        is_busy = state["busy"] != "idle" or state["remaining_steps"] > 0

        if is_busy:
            # While busy, wait is the only sensible action
            if at == ActionType.WAIT:
                score = 0.9
            elif at == ActionType.ABORT_TASK:
                score = 0.2
            else:
                score = -0.1
        else:
            if at == ActionType.CAPTURE_IMAGE:
                if state["requests"]:
                    score = 0.7
                    if action.target_id and any(
                        r["id"] == action.target_id for r in state["requests"]
                    ):
                        score = 0.9
                else:
                    score = -0.3  # No targets available

            elif at == ActionType.COMPRESS_DATA:
                score = 0.9 if state["raw_data"] > 0 else -0.5

            elif at == ActionType.DOWNLINK_TO_STATION:
                if state["gs_visible"] and state["compressed_data"] > 0:
                    score = 0.9
                elif not state["gs_visible"]:
                    score = 0.1
                else:
                    score = 0.2

            elif at == ActionType.SUN_POINT_FOR_CHARGING:
                if state["sunlit"] and state["battery"] < 40:
                    score = 0.9
                elif state["sunlit"] and state["battery"] < 70:
                    score = 0.5
                elif not state["sunlit"]:
                    score = -0.5
                else:
                    score = -0.7  # Battery already high

            elif at == ActionType.WAIT:
                # Penalize if there's something useful to do
                if state["raw_data"] > 0:
                    score = 0.2
                elif state["gs_visible"] and state["compressed_data"] > 0:
                    score = 0.1
                elif state["requests"] and state["battery"] > 20:
                    score = 0.2
                else:
                    score = 0.6

            else:
                score = 0.3

        rewards.append(score)
    return rewards


# ── Reward 3: Priority awareness ─────────────────────────────────
PRIORITY_SCORE = {"high": 1.0, "medium": 0.5, "low": 0.3}

def reward_priority(completions, prompts, **kwargs):
    """Reward targeting high-priority requests."""
    rewards = []
    for c, p in zip(completions, prompts):
        text = _get_completion_text(c).strip()
        action = parse_action_from_text(text)
        state = _parse_state(p)

        if action.action_type == ActionType.CAPTURE_IMAGE:
            if action.target_id and state["requests"]:
                target = next(
                    (r for r in state["requests"] if r["id"] == action.target_id),
                    None,
                )
                if target:
                    score = PRIORITY_SCORE.get(target["priority"], 0.3)
                    # Bonus for picking the highest-priority target
                    max_prio = max(
                        PRIORITY_SCORE.get(r["priority"], 0) for r in state["requests"]
                    )
                    if PRIORITY_SCORE.get(target["priority"], 0) >= max_prio:
                        score = min(1.0, score + 0.2)
                else:
                    score = -0.5  # Invalid target ID
            elif not state["requests"]:
                score = -0.1
            else:
                score = -0.3  # capture without target_id
        else:
            score = 0.5  # Neutral for non-capture actions

        rewards.append(score)
    return rewards


print("Rule-based reward functions ready (no environment needed)")
```

```python
from trl import GRPOConfig

training_args = GRPOConfig(
    temperature=1.0,
    learning_rate=5e-6,
    weight_decay=0.01,
    warmup_ratio=0.1,
    lr_scheduler_type="linear",
    optim="adamw_8bit",
    logging_steps=1,
    per_device_train_batch_size=4,
    gradient_accumulation_steps=4,
    num_generations=4,
    max_prompt_length=1000,
    max_completion_length=20,
    num_train_epochs=5,
    save_steps=100,
    report_to="none",
    output_dir="outputs",
    save_total_limit=1,
)
```

```python
from trl import GRPOTrainer

trainer = GRPOTrainer(
    model=model,
    processing_class=tokenizer,
    reward_funcs=[reward_format, reward_context, reward_priority],
    train_dataset=dataset,
    args=training_args,
)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

logs = pd.DataFrame(trainer.state.log_history)

def smooth(y, alpha=0.15):
    """Exponential moving average — no edge artifacts."""
    s = np.empty_like(y, dtype=float)
    s[0] = y[0]
    for i in range(1, len(y)):
        s[i] = alpha * y[i] + (1 - alpha) * s[i - 1]
    return s

# ── Individual reward plots ───────────────────────────────────────
reward_cols = {
    "rewards/reward_format/mean":   "Expert Match",
    "rewards/reward_priority/mean": "Priority Awareness",
    "rewards/reward_context/mean":  "Context Appropriateness",
}
available = {k: v for k, v in reward_cols.items() if k in logs.columns}

if not available:
    print(f"No reward columns found. Available: {list(logs.columns)}")
else:
    df = logs.dropna(subset=list(available.keys()), how="all").copy()
    x = df["step"].values
    items = list(available.items())
    colors = ["#2196F3", "#4CAF50", "#FF9800"]

    # Row 1: first two reward plots
    fig, axes = plt.subplots(1, 2, figsize=(14, 4))
    for idx in range(min(2, len(items))):
        col, label = items[idx]
        y = df[col].values
        axes[idx].plot(x, y, alpha=0.25, color=colors[idx], linewidth=0.8)
        axes[idx].plot(x, smooth(y), color=colors[idx], linewidth=2, label=f"{label} (smoothed)")
        axes[idx].set_xlabel("Step")
        axes[idx].set_ylabel("Reward")
        axes[idx].set_title(label)
        axes[idx].legend(loc="lower right", fontsize=9)
        axes[idx].grid(alpha=0.3)
    fig.suptitle("Reward Curves", fontsize=14)
    plt.tight_layout()
    plt.show()

    # Row 2: third reward plot
    if len(items) > 2:
        col, label = items[2]
        y = df[col].values
        fig, ax = plt.subplots(figsize=(7, 4))
        ax.plot(x, y, alpha=0.25, color=colors[2], linewidth=0.8)
        ax.plot(x, smooth(y), color=colors[2], linewidth=2, label=f"{label} (smoothed)")
        ax.set_xlabel("Step")
        ax.set_ylabel("Reward")
        ax.set_title(label)
        ax.legend(loc="lower right", fontsize=9)
        ax.grid(alpha=0.3)
        plt.tight_layout()
        plt.show()

# ── Training loss plot ────────────────────────────────────────────
if "loss" in logs.columns:
    loss_df = logs.dropna(subset=["loss"]).copy()
    x_loss = loss_df["step"].values
    y_loss = loss_df["loss"].values

    fig, ax = plt.subplots(figsize=(10, 4))
    ax.plot(x_loss, y_loss, alpha=0.25, color="#E91E63", linewidth=0.8)
    ax.plot(x_loss, smooth(y_loss), color="#E91E63", linewidth=2, label="Loss (smoothed)")
    ax.set_xlabel("Step")
    ax.set_ylabel("Loss")
    ax.set_title("Training Loss")
    ax.legend(fontsize=9)
    ax.grid(alpha=0.3)
    plt.tight_layout()
    plt.show()
else:
    print(f"No 'loss' column found. Available: {list(logs.columns)}")
```
