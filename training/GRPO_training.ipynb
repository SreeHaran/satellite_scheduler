```python
import textwrap

SYSTEM_PROMPT = textwrap.dedent(
    """
    You are an AI agent controlling a satellite for mission-critical imaging and data downlink. You will Capture priority targets, compress data, downlink to ground station, and manage battery.

    Mission objective:
    - maximize completed mission value
    - prefer high-priority requests
    - capture useful images before deadlines
    - compress raw data before downlink
    - downlink compressed data during ground-station visibility
    - preserve battery and avoid storage overflow

    Allowed actions:
    1. wait
    2. abort_task
    3. sun_point_for_charging
    4. compress_data
    5. downlink_to_station
    6. capture_image

    Important environment facts:
    - capture_image requires a target_id
    - capture creates RAW data
    - compress_data converts RAW data into COMPRESSED data and frees storage
    - downlink_to_station only helps when ground station is visible and compressed data exists
    - sun_point_for_charging only helps when sunlit
    - if the satellite is already busy, any non-abort action continues the current task

    Response rules:
    - output exactly one JSON object
    - no markdown
    - no prose
    - schema:
        - For actions without parameters: "wait", "abort_task", "sun_point_for_charging", "compress_data", "downlink_to_station"
        - For capture_image action: "capture_image <target-id>". Example: "capture_image 42"
    """
).strip()
```

```python
from typing import Any, Dict, List, Optional, Tuple

def build_user_prompt(
    step: int,
    obs: SatelliteSchedulerObservation,
    last_reward: float,
    history: List[str],
) -> str:
    # Extract all state attributes from observation
    current_time = obs.current_time
    attitude = obs.attitude
    busy_status = obs.busy_status
    remaining_action_steps = obs.remaining_action_steps
    battery_level = obs.battery_level
    storage_used = obs.storage_used
    raw_data_amount = obs.raw_data_amount
    compressed_data_amount = obs.compressed_data_amount
    sunlit_status = obs.sunlit_status
    ground_station_visible = obs.ground_station_visible
    pending_request_queue = obs.pending_request_queue
    current_selected_request_id = obs.current_selected_request_id

    sunlit_str = "SUNLIT" if sunlit_status else "DARK"
    gs_str = "VISIBLE" if ground_station_visible else "NOT VISIBLE"

    try:
        if pending_request_queue:
            queue_lines = "\n".join(
                f"  - id={r.request_id} priority={r.priority.value} mode={r.imaging_mode.value} deadline={r.deadline}s"
                for r in pending_request_queue
            )
            queue_str = f"{len(pending_request_queue)} pending:\n{queue_lines}"
        else:
            queue_str = "none"
    except Exception as e:
        queue_str = "(error reading queue)"

    return textwrap.dedent(
        f"""
        SATELLITE STATE
        ===============
        Step: {step}/{MAX_STEPS}
        
        ORBITAL STATE
        Time: {current_time}s | Attitude: {attitude} | Busy Status: {busy_status}
        Remaining Action Steps: {remaining_action_steps}
        
        RESOURCE LEVELS
        Battery: {battery_level:.1f}% | Storage Used: {storage_used:.1f} GB
        Raw Data: {raw_data_amount:.1f} GB | Compressed Data: {compressed_data_amount:.1f} GB
        
        VISIBILITY
        Sunlit: {sunlit_str} | Ground Station: {gs_str}
        
        MISSION REQUESTS
        {queue_str}
        Currently Selected Request: {current_selected_request_id if current_selected_request_id else "None"}
        
        PERFORMANCE
        Last Reward: {last_reward:.3f}
        
        RECENT ACTIONS
        {chr(10).join(history[-3:]) if history else "(none)"}
        
        What action should the satellite take next?
    """
    ).strip()

def validate_action(action: Dict[str, Any]) -> Dict[str, Any]:
    allowed = {
        "wait",
        "abort_task",
        "sun_point_for_charging",
        "compress_data",
        "downlink_to_station",
        "capture_image",
    }

    action_type = action.get("action_type")
    if action_type not in allowed:
        raise ValueError(f"Invalid action_type: {action_type}")

    if action_type == "capture_image":
        target_id = action.get("target_id")
        if target_id is None or str(target_id).strip() == "":
            raise ValueError("capture_image requires target_id")
        return {"action_type": "capture_image", "target_id": str(target_id)}

    return {"action_type": action_type}


def parse_action(action: Dict[str, Any]) -> Optional[SatelliteSchedulerAction]:
    """Parse LLM response into a SatelliteSchedulerAction."""
    try:
        text_lower = action.action_type

        # Handle capture_image first: extract target_id from response (e.g. "capture_image 42")
        if action_type == "capture_image":
            match = re.search(r"\d+", text)
            target_id = int(match.group()) if match else None
            return SatelliteSchedulerAction(
                action_type=ActionType.CAPTURE_IMAGE, target_id=target_id
            )

        action_mappings = {
            "wait": ActionType.WAIT,
            "abort": ActionType.ABORT_TASK,
            "sun_point": ActionType.SUN_POINT_FOR_CHARGING,
            "charge": ActionType.SUN_POINT_FOR_CHARGING,
            "compress": ActionType.COMPRESS_DATA,
            "downlink": ActionType.DOWNLINK_TO_STATION,
        }

        for keyword, action_type in action_mappings.items():
            if keyword in text_lower:
                return SatelliteSchedulerAction(action_type=action_type)

        # Default to wait if unclear
        return SatelliteSchedulerAction(action_type=ActionType.WAIT)
    except Exception as e:
        return SatelliteSchedulerAction(action_type=ActionType.WAIT)
```

```python
from datasets import Dataset


def build_grpo_dataset(
    size: int = 10,
    mix_tasks: bool = True,
) -> Dataset:
    """
    Minimal prompt-only dataset for GRPO.
    GRPO gets the real supervision from the environment rollout,
    so this dataset only needs task prompts.
    """

    prompts = """ You are an AI agent controlling a satellite for mission-critical imaging and data downlink.
    Your goal: Capture priority targets, compress data, downlink to ground station, and manage battery.
    
    Available actions: wait, abort_task, sun_point_for_charging, compress_data, downlink_to_station, capture_image
    
    Strategy tips:
        "Operate the satellite to complete as many high-priority imaging requests as possible.",
        "Choose actions that maximize mission completion for urgent requests.",
        "Prioritize high-value captures while avoiding obviously wasteful actions.",
        "Plan satellite operations end-to-end: charging, capture, compression, and downlink.",
        "Balance mission value, battery safety, storage safety, and request deadlines.",
        "Act like a smart onboard scheduler that maximizes total episode performance.",
        "Manage storage carefully by balancing capture, compression, and downlink.",
        "Optimize data flow: capture useful data, compress it, and downlink efficiently.",
        "Avoid storage saturation while still completing imaging tasks.",
    
    Respond with a single action name and optional parameters. Note that optional parameters (like target_id) should only be included for the capture_image action.
    The target id for the capture_image action should be selected from the pending request queue, prioritizing higher priority targets.
    Output format:
    - For actions without parameters: "wait", "abort_task", "sun_point_for_charging", "compress_data", "downlink_to_station"
    - For capture_image action: "capture_image <target-id>". Example: "capture_image 42"
    DO NOT GIVE ANYTHING EXTRA APART FROM THE REQUIRED OUTPUT. ONLY THOSE OUTPUTS ARE ENOUGH. END THE ANSWER AFTER THAT
    """

    return Dataset.from_dict({"prompt": [prompts] * size})
```

```python
BASE_URL= "https://sreeharan-satellite-scheduler.hf.space"
MAX_STEPS = 180
TEMPERATURE = 0.5
MAX_TOKENS = 50
SUCCESS_SCORE_THRESHOLD = 0.5
```

```python
import asyncio
import re
import torch

try:
    import nest_asyncio
    nest_asyncio.apply()
except ImportError:
    get_ipython().system('pip -q install nest_asyncio')
    import nest_asyncio
    nest_asyncio.apply()

# Shared storage: rollout_func writes, reward functions read
_rollout_rewards = {"easy": [], "medium": [], "hard": []}


def parse_action_from_text(text: str) -> SatelliteSchedulerAction:
    """Parse raw LLM text into a SatelliteSchedulerAction."""
    try:
        text_lower = text.lower().strip()

        if "capture" in text_lower or "image" in text_lower:
            match = re.search(r"\d+", text)
            target_id = int(match.group()) if match else None
            return SatelliteSchedulerAction(
                action_type=ActionType.CAPTURE_IMAGE, target_id=target_id
            )

        action_mappings = {
            "wait": ActionType.WAIT,
            "abort": ActionType.ABORT_TASK,
            "sun_point": ActionType.SUN_POINT_FOR_CHARGING,
            "charge": ActionType.SUN_POINT_FOR_CHARGING,
            "compress": ActionType.COMPRESS_DATA,
            "downlink": ActionType.DOWNLINK_TO_STATION,
        }

        for keyword, action_type in action_mappings.items():
            if keyword in text_lower:
                return SatelliteSchedulerAction(action_type=action_type)

        return SatelliteSchedulerAction(action_type=ActionType.WAIT)
    except Exception:
        return SatelliteSchedulerAction(action_type=ActionType.WAIT)


async def _run_episode(mdl, tok):
    """Run one full environment episode. Returns (action_texts, grades)."""
    env = SatelliteSchedulerEnv(base_url=BASE_URL)
    try:
        result = await env.reset()
        history, actions = [], []
        last_reward = 0.01

        for step in range(1, MAX_STEPS + 1):
            if result.done:
                break

            obs = result.observation
            user_prompt = build_user_prompt(step, obs, last_reward, history)

            messages = [
                {"role": "system", "content": SYSTEM_PROMPT},
                {"role": "user", "content": user_prompt},
            ]

            inputs = tok.apply_chat_template(
                messages, return_tensors="pt", add_generation_prompt=True
            ).to(mdl.device)

            with torch.no_grad():
                out = mdl.generate(
                    input_ids=inputs,
                    max_new_tokens=MAX_TOKENS,
                    temperature=TEMPERATURE,
                    do_sample=True,
                )

            response = tok.decode(
                out[0][inputs.shape[1]:], skip_special_tokens=True
            ).strip()

            action = parse_action_from_text(response)
            result = await env.step(action)

            reward = result.reward or 0.0
            last_reward = reward
            actions.append(response)
            history.append(
                f"Step {step}: {action.action_type.name} (reward={reward:+.3f})"
            )

        state = await env.state()
        grades = grade_all(state.episode_stats)
        return actions, grades
    finally:
        await env.close()


def rollout_func(prompts, trainer):
    """
    GRPO rollout function. Runs one env episode per prompt, stores grades in
    the global _rollout_rewards dict so reward functions can read them directly.
    """
    global _rollout_rewards
    loop = asyncio.get_event_loop()

    FastLanguageModel.for_inference(model)

    prompt_ids_list = []
    completion_ids_list = []
    easy_rewards = []
    medium_rewards = []
    hard_rewards = []

    for p_idx, prompt in enumerate(prompts):
        if isinstance(prompt, str):
            p_ids = tokenizer.encode(prompt)
        else:
            p_ids = tokenizer.apply_chat_template(
                prompt, add_generation_prompt=True
            )

        actions, grades = loop.run_until_complete(
            _run_episode(model, tokenizer)
        )

        completion_text = "\n".join(actions) if actions else "wait"
        c_ids = tokenizer.encode(completion_text, add_special_tokens=False)

        prompt_ids_list.append(p_ids)
        completion_ids_list.append(c_ids)
        easy_rewards.append(grades["easy"])
        medium_rewards.append(grades["medium"])
        hard_rewards.append(grades["hard"])

        print(
            f"  [rollout] prompt {p_idx}: "
            f"easy={grades['easy']:.3f} mid={grades['medium']:.3f} hard={grades['hard']:.3f} "
            f"({len(actions)} steps)"
        )

    # Store rewards globally so reward_easy/mid/hard can read them
    _rollout_rewards = {
        "easy": easy_rewards,
        "medium": medium_rewards,
        "hard": hard_rewards,
    }

    FastLanguageModel.for_training(model)

    return {
        "prompt_ids": prompt_ids_list,
        "completion_ids": completion_ids_list,
        "logprobs": None,
    }
```

```python
from trl import GRPOConfig

OUTPUT_DIR = "Satellite-model"

training_args = GRPOConfig(
    temperature=1.0,
    learning_rate=5e-5,
    weight_decay=0.01,
    warmup_ratio=0.1,
    lr_scheduler_type="linear",
    optim="adamw_8bit",
    logging_steps=1,
    per_device_train_batch_size=2,
    gradient_accumulation_steps=1,
    num_generations=2,
    max_prompt_length=1000,
    max_completion_length=50,
    num_train_epochs=10,
    save_steps=100,
    report_to="none",
    output_dir="outputs",
)
```

```python
import asyncio
import re
import torch
from concurrent.futures import ThreadPoolExecutor

try:
    import nest_asyncio
    nest_asyncio.apply()
except ImportError:
    get_ipython().system('pip -q install nest_asyncio')
    import nest_asyncio
    nest_asyncio.apply()


def parse_action_from_text(text: str) -> SatelliteSchedulerAction:
    try:
        text_lower = text.lower().strip()
        if "capture" in text_lower or "image" in text_lower:
            match = re.search(r"\d+", text)
            target_id = int(match.group()) if match else None
            return SatelliteSchedulerAction(
                action_type=ActionType.CAPTURE_IMAGE, target_id=target_id
            )
        for kw, at in {
            "wait": ActionType.WAIT, "abort": ActionType.ABORT_TASK,
            "sun_point": ActionType.SUN_POINT_FOR_CHARGING, "charge": ActionType.SUN_POINT_FOR_CHARGING,
            "compress": ActionType.COMPRESS_DATA, "downlink": ActionType.DOWNLINK_TO_STATION,
        }.items():
            if kw in text_lower:
                return SatelliteSchedulerAction(action_type=at)
        return SatelliteSchedulerAction(action_type=ActionType.WAIT)
    except Exception:
        return SatelliteSchedulerAction(action_type=ActionType.WAIT)


async def _run_episode(mdl, tok):
    env = SatelliteSchedulerEnv(base_url=BASE_URL)
    try:
        result = await env.reset()
        history, last_reward = [], 0.01
        for step in range(1, MAX_STEPS + 1):
            if result.done:
                break
            obs = result.observation
            user_prompt = build_user_prompt(step, obs, last_reward, history)
            messages = [
                {"role": "system", "content": SYSTEM_PROMPT},
                {"role": "user", "content": user_prompt},
            ]
            inputs = tok.apply_chat_template(
                messages, return_tensors="pt", add_generation_prompt=True
            ).to(mdl.device)
            with torch.no_grad():
                out = mdl.generate(
                    input_ids=inputs, max_new_tokens=MAX_TOKENS,
                    max_length=None,
                    temperature=TEMPERATURE, do_sample=True,
                )
            response = tok.decode(out[0][inputs.shape[1]:], skip_special_tokens=True).strip()
            action = parse_action_from_text(response)
            result = await env.step(action)
            reward = result.reward or 0.0
            last_reward = reward
            history.append(f"Step {step}: {action.action_type.name} (reward={reward:+.3f})")
        state = await env.state()
        return grade_all(state.episode_stats)
    finally:
        await env.close()


# ── Dual-GPU setup ──────────────────────────────────────────────
_num_gpus = torch.cuda.device_count()
_gpu_models = [(model, tokenizer)]  # GPU 0 = training model

if _num_gpus >= 2:
    try:
        print(f"{_num_gpus} GPUs detected — loading inference copy on cuda:1 ...")
        _m2, _t2 = FastLanguageModel.from_pretrained(
            model_name=model_path, max_seq_length=2048,
            dtype=None, load_in_4bit=True, device_map={"": 1},
        )
        _m2.load_adapter(model_path, adapter_name="default")
        _gpu_models.append((_m2, _t2))
        print("Dual-GPU episodes enabled")
    except Exception as e:
        print(f"Single-GPU fallback (cuda:1 load failed: {e})")
else:
    print(f"Single GPU mode ({_num_gpus} GPU)")


def _sync_lora_to_gpu1():
    """Copy LoRA weights from GPU 0 training model to GPU 1 inference model."""
    if len(_gpu_models) < 2:
        return
    with torch.no_grad():
        src = dict(model.named_parameters())
        for name, p in _gpu_models[1][0].named_parameters():
            if p.requires_grad and name in src:
                p.data.copy_(src[name].data.to(p.device))


def _episode_worker(args):
    """Thread worker — runs one episode on the assigned GPU."""
    idx, mdl, tok = args
    loop = asyncio.new_event_loop()
    asyncio.set_event_loop(loop)
    try:
        return loop.run_until_complete(_run_episode(mdl, tok))
    finally:
        loop.close()


class _BatchGrader:
    """
    Runs env episodes ONCE per batch, in parallel across available GPUs.
    Uses a call counter (not batch size) so consecutive same-size batches
    don't return stale cached results.
    """
    def __init__(self, num_fns=3):
        self._grades = []
        self._reads = 0
        self._num_fns = num_fns

    def get(self, completions, key):
        if self._reads >= self._num_fns:
            self._reads = 0

        if self._reads == 0:
            n = len(completions)
            # Sync updated LoRA weights to GPU 1 before episodes
            _sync_lora_to_gpu1()
            for mdl, _ in _gpu_models:
                FastLanguageModel.for_inference(mdl)

            # Build tasks — round-robin across GPUs
            tasks = [(i, *_gpu_models[i % len(_gpu_models)]) for i in range(n)]
            # 2 workers per GPU: overlap HTTP I/O with GPU compute
            workers = min(n, len(_gpu_models) * 2)

            with ThreadPoolExecutor(max_workers=workers) as pool:
                self._grades = list(pool.map(_episode_worker, tasks))

            FastLanguageModel.for_training(model)
            for i, g in enumerate(self._grades):
                print(f"  [ep {i}] easy={g['easy']:.4f} mid={g['medium']:.4f} hard={g['hard']:.4f}")

        self._reads += 1
        return [float(g[key]) for g in self._grades]


grader = _BatchGrader(num_fns=3)
```

```python
from trl import GRPOTrainer


def reward_easy(completions, **kwargs):
    return grader.get(completions, "easy")

def reward_mid(completions, **kwargs):
    return grader.get(completions, "medium")

def reward_hard(completions, **kwargs):
    return grader.get(completions, "hard")


trainer = GRPOTrainer(
    model=model,
    processing_class=tokenizer,
    reward_funcs=[reward_easy, reward_mid, reward_hard],
    train_dataset=dataset,
    args=training_args,
)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Logs from GRPO trainer
logs = pd.DataFrame(trainer.state.log_history)

# Keep rows where these rewards exist
reward_cols = ["rewards/reward_easy/mean", "rewards/reward_mid/mean", "rewards/reward_hard/mean"]
available = [c for c in reward_cols if c in logs.columns]
if len(available) < 3:
    raise ValueError(f"Missing reward columns. Found: {available}. Available columns: {list(logs.columns)}")

df = logs.dropna(subset=available, how="all").copy()

# X-axis: prefer epoch, fallback to step
x_col = "epoch" if "epoch" in df.columns else "step"
if x_col == "epoch":
    # smooth jitter by averaging repeated epoch logs
    plot_df = df.groupby(df["epoch"].round(3))[available].mean().reset_index()
else:
    plot_df = df[["step"] + available].copy()

plt.figure(figsize=(10, 5))
for col in available:
    plt.plot(plot_df[x_col], plot_df[col], marker="o", linewidth=1.8, label=col)

plt.xlabel("Epoch" if x_col == "epoch" else "Step")
plt.ylabel("Reward")
plt.title("Easy / Medium / Hard Rewards")
plt.grid(alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()
```
